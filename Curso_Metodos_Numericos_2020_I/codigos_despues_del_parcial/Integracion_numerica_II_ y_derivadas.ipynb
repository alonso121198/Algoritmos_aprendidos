# Integral evaluacion en cualquier intervalo

```python
# definimos la ecuacion gauss que evalua la integral con nodos ya sea de normal , lobatto o radau
# en todas sera con la funcion de peso "legendre"
def gaussQuad(f=None,a=0,b=0,m=1,tipo="normal"):
    '''
    I = gaussQuad lo que hace es evaluar la integral con cuadratura de gauss de nodos
    f = funcion a integrar 
    a , b = limites de integracion inferior y superior respectivamente
    m = cuantos puntos nodos se usara
    tipo = que tipo de cuadratura ("normal" , "radau" ,"lobato")
    '''
    c1 = (b+a)/2.0
    c2 = (b-a)/2.0
    if tipo == "normal":
        x,A = gausslegendreNodes(m)
    elif tipo == "radau":
        x,A = gaussradaulegendreNodes(m)
    elif tipo == "lobatto":
        x,A = gausslobattolegendreNodes(m)
    
    suma = 0 
    for i in range(len(x)):
        suma = suma + A[i]*f(c1 + c2*x[i])
        
    return c2*suma
```

```python
# ejemplo del profesor para resolver 
def funcionprueba(x):
    return np.float64( np.cos(x) )

# hagamos con gauss-legendre de los 3 tipos
I1 = gaussQuad(f=funcionprueba,a=0,b=np.pi/2,m=3,tipo="normal")
I2 = gaussQuad(f=funcionprueba,a=0,b=np.pi/2,m=7,tipo="radau")
I3 = gaussQuad(f=funcionprueba,a=0,b=np.pi/2,m=3,tipo="lobatto")
valor_analitco = np.sin(np.pi/2) - np.sin(0)


print(f"Con gauss-legendre : {I1}")
print(f"Con gauss-radau-legendre : {I2}")
print(f"Con gauss-lobatto-legendre : {I3}")
print(f"Valor analitico es : {valor_analitco}")
```

```python
def differentation(func=None, evalf=None, orden=1, h=None, tipo="central", condata=False, data=None):
    '''
    Funcion que si es sin data te pide ( condata = False )
    - funcion = que quieres diferenciar
    - evalf = punto de evaluacion
    - orden = orden de la derivada ( primera , segunda , tercera , cuarta )
    - tipo = "forward" , "backward" , "central" son las opciones

    Si la funcion es con data entonces se acepta solo la data y el orden como argumentos ( condata = True )
    - data = data donde la primera fila son los valores x y la segunda fila son los y correspondientes
    - orden = orden de la derivada ( primera , segunda , tercera , cuarta )
    '''
    if condata:

        ######### Forma de la data #######
        # - esperamos una data de la siguiente manera , tenga los puntos "x" y "y" donde se ha hecho las mediciones
        # lo que retornaremos sera la derivada en esos puntos medidos

        # necesitamos el tama√±o de la data
        n = data.shape[1]
        h = data[0, 1] - data[0, 0]
        derivada = []  # sera la lista que retornaremos de las derivadas
        # tenemos que aplicar al inicio forward , en el medio "central" y al extremo derecho "backward"
        if orden == 1:
            if n < 3:
                print("Tu data es muy escasa")
                return None

                # ------------- Inicio de data -----------#
            # se aplicara "forward"
            # //// deriv = (-3)*func(x) + 4*func(x+h) + (-1)*func(x+2*h) ///
            deriv = (-3) * data[1, 0] + 4 * data[1, 1] + (-1) * data[1, 2]
            deriv = deriv / (2 * h)
            derivada.append(deriv)
            # ------------- central de data --------- #
            # se aplicara "central"
            for i in range(1, n - 1):
                # /// deriv = (-1)*func(x-h) + 0*func(x) + 1*func(x+h) ///
                deriv = (-1) * data[1, i - 1] + 0 * data[1, i] + 1 * data[1, i + 1]
                deriv = deriv / (2 * h)
                derivada.append(deriv)
            # ------------final de data ------------- #
            # se aplicara "backward"
            # /// deriv = 1*unc(x-2*h) + (-4)*func(x-h) + 3*func(x) ///
            deriv = 1 * data[1, n - 1 - 2] + (-4) * data[1, n - 1 - 1] + 3 * data[1, n - 1]
            deriv = deriv / (2 * h)
            derivada.append(deriv)

        if orden == 2:

            if n < 4:
                print("Tu data es muy escasa")
                return None

            # ------------- Inicio de data -----------#
            # se aplicara "forward"
            # //// deriv = 2*func(x) + (-5)*func(x+h) + 4*func(x+2*h) + (-1)*func(x+3*h) ////
            deriv = 2 * data[1, 0] + (-5) * data[1, 1] + 4 * data[1, 2] + (-1) * data[1, 3]
            deriv = deriv / (2 * h)
            derivada.append(deriv)
            # ------------- central de data --------- #
            # se aplicara "central"
            for i in range(1, n - 1):
                # //// deriv = 1*func(x-h) + (-2)*func(x) + 1*func(x+h)///
                deriv = 1 * data[1, i - 1] + (-2) * data[1, i] + 1 * data[1, i + 1]
                deriv = deriv / (2 * h)
                derivada.append(deriv)
            # ------------final de data ------------- #
            # se aplicara "backward"
            # //// deriv = (-1)*func(x-3*h) + 4*unc(x-2*h) + (-5)*func(x-h) + 2*func(x) ///
            deriv = (-1) * data[1, n - 1 - 3] + 4 * data[1, n - 1 - 2] + (-5) * data[1, n - 1 - 1] + 2 * data[1, n - 1]
            deriv = deriv / (2 * h)
            derivada.append(deriv)

        if orden == 3:
            # pendiente , aca tiene otra forma de evaluar la data
            pass
        if orden == 4:
            # pendiente , aca tiene otra forma de evaluar la data
            pass

        return np.array(derivada)
    else:
        # usamos el x para hacer mas entendible el codigo
        x = evalf
        # segun el orden aplicara el orden ( primera , segunda ,tercera ) de derivada
        # tambien si es backward , central o forward
        # aca hemos supuesto que la funcion se puede evaluar en donde sea osea x+algo siempre existe
        if orden == 1:
            if tipo == "backward":
                derivada = 0 * func(x - 5 * h) + 0 * func(x - 4 * h) + 0 * func(x - 3 * h) + 1 * func(x - 2 * h) + (
                    -4) * func(x - h) + 3 * func(x)
            elif tipo == "central":
                derivada = 0 * func(x - 2 * h) + (-1) * func(x - h) + 0 * func(x) + 1 * func(x + h) + 0 * func(
                    x + 2 * h)
            elif tipo == "forward":
                derivada = (-3) * func(x) + 4 * func(x + h) + (-1) * func(x + 2 * h) + 0 * func(x + 3 * h) + 0 * func(
                    x + 4 * h) + 0 * func(x + 5 * h)
            derivada = derivada / (2 * h)

        elif orden == 2:
            if tipo == "backward":
                derivada = 0 * func(x - 5 * h) + 0 * func(x - 4 * h) + (-1) * func(x - 3 * h) + 4 * func(x - 2 * h) + (
                    -5) * func(x - h) + 2 * func(x)
            elif tipo == "central":
                derivada = 0 * func(x - 2 * h) + 1 * func(x - h) + (-2) * func(x) + 1 * func(x + h) + 0 * func(
                    x + 2 * h)
            elif tipo == "forward":
                derivada = 2 * func(x) + (-5) * func(x + h) + 4 * func(x + 2 * h) + (-1) * func(x + 3 * h) + 0 * func(
                    x + 4 * h) + 0 * func(x + 5 * h)
            derivada = derivada / (h ** 2)

        elif orden == 3:
            if tipo == "backward":
                derivada = 0 * func(x - 5 * h) + 3 * func(x - 4 * h) + (-14) * func(x - 3 * h) + 24 * func(x - 2 * h) + (
                    -18) * func(x - h) + 5 * func(x)
            elif tipo == "central":
                derivada = (-1) * func(x - 2 * h) + 2 * func(x - h) + 0 * func(x) + (-2) * func(x + h) + 1 * func(
                    x + 2 * h)
            elif tipo == "forward":
                derivada = (-5) * func(x) + 18 * func(x + h) + (-24) * func(x + 2 * h) + 14 * func(x + 3 * h) + (
                    -3) * func(x + 4 * h) + 0 * func(x + 5 * h)
            derivada = derivada / (2 * h ** 3)
        elif orden == 4:
            if tipo == "backward":
                derivada = (-2) * func(x - 5 * h) + 11 * func(x - 4 * h) + (-24) * func(x - 3 * h) + 26 * func(
                    x - 2 * h) + (-14) * func(x - h) + 3 * func(x)
            elif tipo == "central":
                derivada = 1 * func(x - 2 * h) + (-4) * func(x - h) + 6 * func(x) + (-4) * func(x + h) + 1 * func(
                    x + 2 * h)
            elif tipo == "forward":
                derivada = 3 * func(x) + (-14) * func(x + h) + 26 * func(x + 2 * h) + (-24) * func(
                    x + 3 * h) + 11 * func(x + 4 * h) + (-2) * func(x + 5 * h)
            derivada = derivada / (h ** 4)

        return derivada
```

```python
print("//////////////////////Ejercicio de derivadas//////////////////")
### Resolvemos un problema del profesor alejandro 
t = np.array([200,202,204,206,208,210],dtype=np.float64)
dato1 = np.array([0.75,0.72,0.70,0.68,0.67,0.66],dtype=np.float64)
dato2 = np.array([5120,5370,5560,5800,6030,6240],dtype=np.float64)

# armamos la data 
theta = np.array([t,dato1])
pos_r = np.array([t,dato2])

# hallamos la primera y segunda derivada respecto al tiempo de posicon "r"
deriv1_pos_r = differentation(condata=True,data=pos_r,orden=1)
deriv2_pos_r = differentation(condata=True,data=pos_r,orden=2)

# hallamos la primera y segunda derivada respecto al tiempo de theta
deriv1_theta = differentation(condata=True,data=theta,orden=1)
deriv2_theta = differentation(condata=True,data=theta,orden=2)

# luego hallando la data velocidad en coordenadas polares // esta data sera de tuplas // usaremos list comprehension
# "zip" lo que hace es que las variables se correspondan de una en una y se itere (esto acorta el trabajo)
data_velocidad_polar = [[derivr,r*derivtheta] for r,derivr,theta,derivtheta in zip(pos_r[1],deriv1_pos_r,theta[1],deriv1_theta)]

# luego hallando la data aceleracion en coordenadas polares // esta data sera de tuplas // usaremos list comprehension
# "zip" lo que hace es que las variables se correspondan de una en una y se itere (esto acorta el trabajo)
data_aceleracion_polar = [[deriv2r-r*derivtheta**2,r*deriv2theta + 2*derivr*derivtheta] for r,derivr,deriv2r,theta,derivtheta,deriv2theta in zip(pos_r[1],deriv1_pos_r,deriv2_pos_r,theta[1],deriv1_theta,deriv2_theta)]

# transformamos los vectores en cartesianos
# Ax = Ar*cos(teta)-Ateta*sen(teta)
# Ay = Ar*sen(teta)-Ateta*cos(teta)

data_velocidad_cartesiano = []
for polar,pos,angulo in zip(data_velocidad_polar,pos_r[1],theta[1]):
    x = polar[0]*np.cos(angulo)-polar[1]*np.sin(angulo)
    y = polar[0]*np.sin(angulo)+polar[1]*np.cos(angulo) 
    data_velocidad_cartesiano.append([x,y])

data_aceleracion_cartesiano = []
for polar,pos,angulo in zip(data_aceleracion_polar,pos_r[1],theta[1]):
    x = polar[0]*np.cos(angulo)-polar[1]*np.sin(angulo)
    y = polar[0]*np.sin(angulo)+polar[1]*np.cos(angulo) 
    data_aceleracion_cartesiano.append([x,y])

# imprimimos nuestros resultados

print("-"*20,"velocidad en polares")
for i in range(0,len(data_velocidad_polar)):
    print(f"Para el tiempo {t[i]} se tiene el vector velocidad polar ({data_velocidad_polar[i][0]}, {data_velocidad_polar[i][1]:.4f})")
print("-"*20,"velocidad en cartesiano")
for i in range(0,len(data_velocidad_cartesiano)):
    print(f"Para el tiempo {t[i]} se tiene el vector velocidad cartesiano ({data_velocidad_cartesiano[i][0]}, {data_velocidad_cartesiano[i][1]})")    
print("-"*20,"aceleracion en polares")
for i in range(0,len(data_aceleracion_polar)):
    print(f"Para el tiempo {t[i]} se tiene el vector velocidad polar ({data_aceleracion_polar[i][0]:.5f}, {data_aceleracion_polar[i][1]:.5f})")
print("-"*20,"aceleracion en cartesiano")
for i in range(0,len(data_aceleracion_cartesiano)):
    print(f"Para el tiempo {t[i]} se tiene el vector velocidad polar ({data_aceleracion_cartesiano[i][0]}, {data_aceleracion_cartesiano[i][1]})")   
# Ahora graficamos los vectores en polares 

# creamos una figura con su eje que almaceara todas las propiedades de la imagen
fig, ax = plt.subplots()
# dibujamos las velocidades
X = pos_r[1]*np.cos(theta[1])
Y = pos_r[1]*np.sin(theta[1])
compvelx = [dato[0] for dato in data_velocidad_cartesiano ]
compvely = [dato[1] for dato in data_velocidad_cartesiano ]
compacelx = [dato[0] for dato in data_aceleracion_cartesiano ]
compacely = [dato[1] for dato in data_aceleracion_cartesiano ]

# usar quiver documentacion (https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.quiver.html#matplotlib.axes.Axes.quiver)
ax.quiver(X,Y,compvelx,compacely,color="r",label="velocidad")
ax.quiver(X,Y,compacelx,compacely,color="b",label="aceleracion")
ax.set_xscale("log")
ax.set_yscale("log")
ax.legend( loc='upper left', shadow=True) # se pone la legenda 
fig.suptitle('Resultado del ejercicio de las derivadas', fontsize=20)
plt.show()

# la posicion de los vectores lo puedo conseguir
# me falta pasar la velocidad a vectores en cartesianos 
```
